```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
file_path = '/content/GoogleDrive/MyDrive/CallCenters.csv'
df = pd.read_csv(file_path)
```

```python
# 1. Total number of calls received by all call centers during the entire period
total_calls = df['calls'].sum()
print(f"1. Total number of calls received by all call centers: {total_calls}")
```

```python
# 2. Total number of calls received regarding Photoshop during the entire period
photoshop_calls = df[df['product'] == 'Photoshop']['calls'].sum()
print(f"2. Total number of calls received regarding Photoshop: {photoshop_calls}")
```

```python
# 3. Product with the most calls in 2020 and the number of calls
max_calls_2020 = df[df['year'] == 2020].groupby('product')['calls'].sum().idxmax()
max_calls_2020_count = df[(df['year'] == 2020) & (df['product'] == max_calls_2020)]['calls'].sum()
print(f"3. Product with the most calls in 2020: {max_calls_2020}, Calls: {max_calls_2020_count}")
```

```python
# 4. Year with the highest number of calls and the total number of calls
max_calls_year = df.groupby('year')['calls'].sum().idxmax()
max_calls_year_count = df.groupby('year')['calls'].sum().max()
print(f"4. Year with the highest number of calls: {max_calls_year}, Total Calls: {max_calls_year_count}")
```

```python
# 5. Center with the highest number of calls
max_calls_center = df.groupby('center')['calls'].sum().idxmax()
print(f"5. Center with the highest number of calls: {max_calls_center}")
```

```python
# 6. Centers in order based on the percentage of closed issues
center_closed_percent = (df.groupby('center')['closed'].sum() / df.groupby('center')['calls'].sum()) * 100
centers_ordered_by_closed_percent = center_closed_percent.sort_values(ascending=False).index
print(f"6. Centers in order based on percentage of closed issues: {list(centers_ordered_by_closed_percent)}")
```

```python
# 7. Display a chart showing the total calls in each center
plt.figure(figsize=(12, 6))
sns.barplot(x='center', y='calls', data=df, estimator=sum, ci=None)
plt.title('Total Calls in Each Center')
plt.xlabel('Center')
plt.ylabel('Total Calls')
plt.show()
```

```python
# 8. Display a chart showing the number of calls related to each product
plt.figure(figsize=(12, 6))
sns.barplot(x='product', y='calls', data=df, estimator=sum, ci=None)
plt.title('Number of Calls Related to Each Product')
plt.xlabel('Product')
plt.ylabel('Total Calls')
plt.show()
```

```python
# 9. Display a chart showing the distribution of calls over the years
plt.figure(figsize=(12, 6))
sns.barplot(x='year', y='calls', data=df, estimator=sum, ci=None)
plt.title('Distribution of Calls Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Calls')
plt.show()
```

```python
# 10. Display a chart showing the distribution of calls across the centers for each year
plt.figure(figsize=(16, 8))
sns.barplot(x='year', y='calls', hue='center', data=df, estimator=sum, ci=None)
plt.title('Distribution of Calls Across Centers for Each Year')
plt.xlabel('Year')
plt.ylabel('Total Calls')
plt.legend(title='Center', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
